## Problem 2: Learn string formatting

Now we will learn the concept of string formatting. Suppose we want to print a message that says:

"The area of a triangle with a base XXX and height XXX is equal to XXX"

We want to replace the values of "XXX" by the correponding variables. To do so, we can use the .format option in python. Here's an example:

```python
age = 35
name = "Vishesh"
print("My name is {0} and I am {1} years old.".format(name,age))
```

* The idea here is that we use the curly braces as placeholders,
* The number inside the curly braces corresponds to the variables in the .format, e.g. 0 correponds to 1st variable, 1 to the second variable and so on.
* Note there is no comma after the closing double quotes in the print statement, but a dot `.`

## Problem 3: Formatting numbers

Next, we will learn how to format numbers. Suppose we want to print a number upto 3 decimal places.

We can use the following code:

```python
pi = 22/7
print("pi rounded to 1 decimal place is: {0:.1f}. But if we round it to 2 decimals, it is {0:.2f}".format(pi))
```

Let us parse this code, speficially the term `{0:.2f}`:
* `{0}` represents the first variable in the format statement (i.e. pi)
* The colon `:` after `0` tells python that we want to specify some formatting options
* `.2f` means that we want to print a number with 2 decimal places. The "f" represents the floating point number
* For e.g. if we want 3 decimal places, we say `.3f` and so on.
* If we want to print a percentage, we say `.2%` , the 2 denotes the number of decimal places we want.

You have a stock that closed at $550$ on Monday, and then closed at $570$ on Tuesday. Calculate its daily return: the daily return is defined as the difference in the closing prices, divided by the closing price the day before.

* Create variables for storing the price on Monday and price on Tuesday.
* Write down the calculation that will print the daily returns.
* Format the output using the .format option to print the answer upto 4 decimal places
* Format the output using the .format option to print the answer as a percentage.

```python
# your code is here

monday_price = 790
tuesday_price = 700
daily_return = (tuesday_price - monday_price) / monday_price
print("The daily return is {:.4f}.".format(daily_return))
print("The daily return is {:.2%}.".format(daily_return))
```

# Problem 7: A very rudimentary sentiment analysis

To give you a flavor of text mining and the problems that we may come across when doing text mining, we will do a very rudimentary sentiment analysis. Note that this is a highly contrived example to illustrate the basic idea. In the real world, we use more powerul machine learning methods to do sentiment analysis.

What is sentiment analysis? Here's an example:



In the above example, the first review has positive sentiment, and the second review has a negative sentiment. How can we make a computer reconigze such sentiments? We will look at a rudimentary way of doing this.

In any review, for each occurence of the words "good" or "great", we will give it a score of +1.
For each occurence of the words "bad" or "terrible" appears, we will give it a score of -1.
If the total score is positive, the review has a positive sentiment, otherwise it has a negative sentiment.

* Find the sentiment of the review in the review.txt file. (Hint: use the count function to count the number of times a word appears. Be careful regarding upper and lower cases, so you dont miss any words).

```python
# your code here

file_name = r"/content/review.txt"

with open(file_name, 'r') as FILE:
  string = FILE.read()
  FILE.close()

string_lowercase = string.lower()

num_good = string_lowercase.count("good")
print("The number of occurences of the word \"good\" (case insensitive) is {}.".format(num_good))

num_great = string_lowercase.count("great")
print("The number of occurences of the word \"great\" (case insensitive) is {}.".format(num_great))

num_bad = string_lowercase.count("bad")
print("The number of occurences of the word \"bad\" (case insensitive) is {}.".format(num_bad))

num_terrible = string_lowercase.count("terrible")
print("The number of occurences of the word \"terrible\" (case insensitive) is {}.".format(num_terrible))

sentiment_score = num_good + num_great - num_bad - num_terrible

if sentiment_score > 0:
  print("The sentiment is POSITIVE.")
elif sentiment_score < 0:
  print("The sentiment is NEGATIVE.")
else:
  print("The sentiment is NEUTRAL.")
```
